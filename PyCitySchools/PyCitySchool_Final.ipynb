```python
#set up directory
import pandas as pd
```

```python
file_1 = 'Resources/schools_complete.csv'
file_2 = 'Resources/students_complete.csv'

school_df = pd.read_csv(file_1)
student_df = pd.read_csv(file_2)
```

```python
merge_df = pd.merge(school_df,student_df,on='school_name',how='outer')
merge_df = merge_df.drop(columns='School ID')
merge_df.head()
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
#Calculate the percentage of students with a passing math score (70 or greater)
#Calculate the percentage of students with a passing reading score (70 or greater)

total_schools = merge_df['school_name'].nunique()
total_students = merge_df['student_name'].count()
total_budget = merge_df['budget'].unique().sum()
avg_math = round(merge_df['math_score'].mean(),6)
avg_reading = round(merge_df['reading_score'].mean(),6)
math1_df= merge_df.loc[merge_df['math_score']>=70]
reading1_df = merge_df.loc[merge_df['reading_score']>=70] 
percent_math = round((math1_df['math_score'].count()/total_students)*100,6)
percent_reading = round((reading1_df['reading_score'].count()/total_students)*100,6)
avg_overall =(percent_math+percent_reading)/2
#print(total_schools,total_students,total_budget,avg_math,avg_reading,percent_math,percent_reading,avg_overall)

summary_df = pd.DataFrame({'Total Schools':[total_schools],
                          'Total Students':[total_students],
                          'Total Budgets':[total_budget],
                          'Average Math Score':[avg_math],
                          'Average Reading Score':[avg_reading],
                          'Average math Score':[avg_math],
                          '% Passing Math':[percent_math],
                          '$ Passing Reading':[percent_reading],
                          'Overall Passing Rate':[avg_overall]})
summary_format = summary_df.style.format({'Total Students':'{:,.0f}','Total Budgets':'${:,.0f}'})
summary_format
```

```python
new_df = merge_df[['school_name','type','size','budget']]
new_df['Per Student Budget'] = new_df['budget']/new_df['size']

new_index_df = new_df.drop_duplicates().set_index(['school_name'])

avg_df = merge_df[['school_name','reading_score','math_score']].groupby(['school_name']).mean()
#avg_df = new3_df.groupby(['school_name']).mean()

school_math_df = math1_df.loc[:,['school_name','math_score']].groupby(['school_name',]).count()
school_reading_df = reading1_df.loc[:,['school_name','reading_score']].groupby(['school_name',]).count()

school_merge1_df = pd.merge(avg_df,school_reading_df,how='outer',on='school_name')
school_merge2_df = pd.merge(school_merge1_df,school_math_df,how='outer', on='school_name')
school_merge3_df = pd.merge(new_index_df,school_merge2_df,how='outer', on='school_name')

school_merge3_df['% Passing Reading'] = round(school_merge3_df['reading_score_y']/school_merge3_df['size']*100,6)
school_merge3_df['% Passing Math'] = round(school_merge3_df['math_score_y']/school_merge3_df['size']*100,6)
school_merge3_df['Overall Passing Rate'] = (school_merge3_df['% Passing Reading']+school_merge3_df['% Passing Math'])/2

school_merge3_df.index.name = None
final_summary_df = school_merge3_df.copy()

final_rename_df = final_summary_df.rename(columns={'type':'School Type',
                           'size':'Total Students',
                           'budget':'Total School Budget',
                           'reading_score_x':'Average Reading Score',
                           'math_score_x':'Average Math Score',
                           })
final_form_df = final_rename_df.drop(columns=['reading_score_y','math_score_y'])
final_form_df['Total School Budget'] = final_form_df['Total School Budget'].map("${:,.2f}".format)
final_form_df['Per Student Budget'] = final_form_df['Per Student Budget'].map("${:,.2f}".format)

final_form_df
```

```python
top_sort_df = final_form_df.sort_values('Overall Passing Rate',ascending = False)
top_sort_df.head(5)
```

```python
top_sort_df = final_form_df.sort_values('Overall Passing Rate',ascending = True)
top_sort_df.head(5)
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ['<$585', '$585-615', '$615-645', '$645-675']


final_rename_df['Spending Ranges (Per Student)'] = pd.cut(final_rename_df['Per Student Budget'], spending_bins, labels=group_names, include_lowest=True)
bin_budget_df = final_rename_df.loc[:,['Average Math Score','Average Reading Score','% Passing Math',
                               '% Passing Reading','Overall Passing Rate','Spending Ranges (Per Student)']].groupby(['Spending Ranges (Per Student)']).mean()
bin_budget_df
```

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

final_rename_df['School Size'] = pd.cut(final_rename_df['Total Students'], size_bins, labels=group_names, include_lowest=True)
bin_size_df = final_rename_df.loc[:,['Average Math Score','Average Reading Score','% Passing Math',
                               '% Passing Reading','Overall Passing Rate','School Size']].groupby(['School Size']).mean()

bin_size_df
```
